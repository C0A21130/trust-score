# ネットワークの中心性

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.karate_club_graph()
plt.figure(figsize=(7, 7))
nx.draw_spring(G, node_size=400, node_color="red", with_labels=True)
```

```python
def print_centrality(v):
    for key in v:
        print("{:2}：{:<5}".format(key, round(v[key], 3)), end=("  " if (key+1)%5 else "\n"))
    print("")
    print("MAX：", np.argmax(list(v.values())))
```

## 次元中心性
- 次数が多い頂点を中心とする

```python
v = nx.degree_centrality(G)
print_centrality(v)
```

## 近傍中心性
- 他の頂点におけるパスで平均距離が短い頂点を中心とする

```python
v = nx.closeness_centrality(G)
print_centrality(v)
```

## 媒介中心性
- ２頂点間を結ぶ経路上にしばしば現れる頂点を中心とする

```python
v = nx.betweenness_centrality(G)
print_centrality(v)
```

## 固有ベクトル中心性
- 他の中心的頂点と隣接する頂点は中心とする

```python
v = nx.eigenvector_centrality(G)
print_centrality(v)
```

## PageRank

```python
v = nx.pagerank(G)
print_centrality(v)
```
