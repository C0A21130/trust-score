## ランダムグラフ
- ランダムグラフ：erdos_renyi_graph(<頂点数>>, <エッジ作成の確率>)

```python
def p1(x):
    return (1**x)*(np.e**(-1))/scipy.special.factorial(x)
def p3(x):
    return (3**x)*(np.e**(-3))/scipy.special.factorial(x)
def p5(x):
    return (5**x)*(np.e**(-5))/scipy.special.factorial(x)
def p10(x):
    return (10**x)*(np.e**(-10))/scipy.special.factorial(x)
```

```python
# 等差数列：np.arage(<スタート>, <エンド>, <間隔>)

plt.figure(figsize=(12, 9))

er01 = nx.erdos_renyi_graph(100, 0.01)
plt.subplot(3, 4, 1)
nx.draw(er01, node_size=10, node_color="red")
plt.subplot(3, 4, 5)
plt.plot(nx.degree_histogram(er01))
plt.subplot(3, 4, 9)
x = np.arange(0, 4, 1)
y = p1(x)
plt.plot(x, y)

er03 = nx.erdos_renyi_graph(100, 0.03)
plt.subplot(3, 4, 2)
nx.draw(er03, node_size=10, node_color="red")
plt.subplot(3, 4, 6)
plt.plot(nx.degree_histogram(er03))
plt.subplot(3, 4, 10)
x = np.arange(0, 10, 1)
y = p3(x)
plt.plot(x, y)

er05 = nx.erdos_renyi_graph(100, 0.05)
plt.subplot(3, 4, 3)
nx.draw(er05, node_size=10, node_color="red")
plt.subplot(3, 4, 7)
plt.plot(nx.degree_histogram(er05))
plt.subplot(3, 4, 11)
x = np.arange(0, 12, 1)
y = p5(x)
plt.plot(x, y)

er1 = nx.erdos_renyi_graph(100, 0.1)
plt.subplot(3, 4, 4)
nx.draw(er1, node_size=10, node_color="red")
plt.subplot(3, 4, 8)
plt.plot(nx.degree_histogram(er1))
plt.subplot(3, 4, 12)
x = np.arange(0, 20, 1)
y = p10(x)
plt.plot(x, y)

print("１列目：グラフ、２列目：次数ヒストグラム、３列目：ポアソン分布")
plt.show()
```

### コンフィギュレーションモデル
- nx.configuration_model(<辺数のリスト>)

```python
plt.figure(figsize=(6, 3))
deg_seq = [0, 1, 1, 2, 2, 2, 2, 3, 3, 4]

plt.subplot(121)
G = nx.configuration_model(deg_seq)
nx.draw_spring(G, node_size=100, node_color="red")

plt.subplot(122)
G = nx.configuration_model(deg_seq)
nx.draw_spring(G, node_size=100, node_color="red")

plt.show()
```

## スケールフリーグラフ
- nx.barabasi_albert_graph(<頂点数>, <頂点を追加したときに張られる辺数>)

```python
plt.figure(figsize=(6, 9))
ba = nx.barabasi_albert_graph(100, 3)

plt.subplot(311)
nx.draw(ba, node_size=10, node_color="red")

plt.subplot(312)
plt.plot(nx.degree_histogram(ba))

plt.subplot(313)
plt.xscale("log")
plt.yscale("log")
plt.grid("both")
plt.plot(nx.degree_histogram(ba))

plt.show()
```

## スモールワールドグラフ
- nx.watts_strogatz_graph(<頂点数>, <初期状態の各頂点の次数>, <辺の一部をランダムに書き換える確率>)

```python
plt.figure(figsize=(9, 3))

plt.subplot(131)
ws0 = nx.watts_strogatz_graph(100, 4, 0)
nx.draw_circular(ws0, node_size=10, node_color="red")

plt.subplot(132)
ws0 = nx.watts_strogatz_graph(100, 4, 0.3)
nx.draw_circular(ws0, node_size=10, node_color="red")

plt.subplot(133)
ws0 = nx.watts_strogatz_graph(100, 4, 1)
nx.draw_circular(ws0, node_size=10, node_color="red")

plt.show()
```
